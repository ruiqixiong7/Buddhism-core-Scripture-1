```python
create_canvas(400, 400)
background(255)
stroke(0)
fill(0)

img = load_image('bd40c191c3ed0e8bbfd1b862f9d19f14.png')
img = img.resize((30,30)).convert('L')
image(img, 0, 0, 250, 400)
spacing = width/img.height
for y in range(img.height):
    for x in range(img.width): #
        push()
        v = img.getpixel((x,y)) # In range 0 - 255
        radius = remap(v, (0, 255), (spacing, 0.0))
        translate(x*spacing+spacing/2, y*spacing+spacing/2)
        circle(0, 0, radius)
        #rect(x, y, 1, 1)
        pop()
show()
```

```python
txt = "Form is emptiness,emptiness is form".replace(' ', '')

create_canvas(500, 500)

translate(center)
text_align(CENTER, CENTER)
text_font("Times New Roman")
text_size(80)
stroke(0)
fill(255)
i = 0
for c in txt:
    angle = remap(i, 0, len(txt), 0, TWO_PI)
    pos = direction(angle-radians(90))*160
    text(c, pos)
    i += 1  # Increment `i` by 1 manually!
save('test.pdf')
show()
```

```python
txt = "All conditioned things are impermanent; " \
"All conditioned things are unsatisfactory; " \
"All phenomena are without self; Nirvana is peace.".replace(' ', '')

create_canvas(600, 600)
background(255)
translate(center)

text_align(CENTER, CENTER)
text_font("Times New Roman")
text_size(30)
stroke(0)
fill(0)


start_radius = 200   
b = -2               
angle_step = 0.3     
sx = 1.5             
sy = 1.0             

for i, c in enumerate(txt):
    theta = i * angle_step          
    r = start_radius + b * theta    
    x = cos(theta - radians(90)) * r
    y = sin(theta - radians(90)) * r

    
    angle = theta + PI/2

    push_matrix()
    translate((x, y))
    rotate(angle)
    scale(sx, sy)   
    text(c, (0, 0))
    pop_matrix()

save('spiral.pdf')
show()
```

```python
txt = "Form is emptiness,emptiness is form".replace(' ', '')

create_canvas(800,800)
background(255)
translate(center)

text_align(CENTER, CENTER)
text_font("Times New Roman")
text_size(40)   
stroke(0)
fill(255)

w = 500  
h = 220   
sx = 1.8  
sy = 1.0  

def rect_path_with_angle(t, w, h):
    p = t * 4
    if p < 1:      
        x = remap(p, 0, 1, -w/2,  w/2)
        y = -h/2
        angle = 0
    elif p < 2:    
        x =  w/2
        y = remap(p, 1, 2, -h/2,  h/2)
        angle = 90
    elif p < 3:    
        x = remap(p, 2, 3,  w/2, -w/2)
        y =  h/2
        angle = 180
    else:          
        x = -w/2
        y = remap(p, 3, 4,  h/2, -h/2)
        angle = -90
    return (x, y, angle)

for i, c in enumerate(txt):
    t = i / len(txt)
    x, y, angle = rect_path_with_angle(t, w, h)
 
    push_matrix()
    translate((x, y))
    scale(5, 2) 
    rotate(angle)
      
    text(c, (0, 0))
    pop_matrix()

save('test.pdf')
show()
```
